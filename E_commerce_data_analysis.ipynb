```python
# Load datasets
orders = pd.read_csv('/content/olist_orders_dataset.csv')
customers = pd.read_csv('/content/olist_customers_dataset.csv')
items = pd.read_csv('/content/olist_order_items_dataset.csv')
products = pd.read_csv('/content/olist_products_dataset.csv')

# Display basic info
print("Orders dataset shape:", orders.shape)
print("Customers dataset shape:", customers.shape)
print("Order Items dataset shape:", items.shape)
print("Products dataset shape:", products.shape)
```

```python
# Preview the first few rows
print("\nORDERS DATA:")
display(orders.head())

print("\nCUSTOMERS DATA:")
display(customers.head())

print("\nORDER ITEMS DATA:")
display(items.head())

print("\nPRODUCTS DATA:")
display(products.head())
```

```python
print("ORDERS COLUMNS:\n", orders.columns.tolist(), "\n")
print("CUSTOMERS COLUMNS:\n", customers.columns.tolist(), "\n")
print("ITEMS COLUMNS:\n", items.columns.tolist(), "\n")
print("PRODUCTS COLUMNS:\n", products.columns.tolist(), "\n")
```

```python
# Check missing values
print("Missing values in Orders:")
print(orders.isnull().sum())
print("\nMissing values in Customers:")
print(customers.isnull().sum())
print("\nMissing values in Items:")
print(items.isnull().sum())
print("\nMissing values in Products:")
print(products.isnull().sum())
```

```python
# Convert date columns to datetime
date_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
                'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

# Check date range
print("Order date range:", orders['order_purchase_timestamp'].min(), "to", orders['order_purchase_timestamp'].max())
```

```python
# Summary stats
items.describe()
```

```python
# Merge orders + customers + items + products
order_items = pd.merge(items, products, on='product_id', how='left')
order_data = pd.merge(order_items, orders, on='order_id', how='left')
order_data = pd.merge(order_data, customers, on='customer_id', how='left')

# Check result
order_data.shape
```

```python
# Create a 'year-month' column from purchase date
order_data['order_month'] = order_data['order_purchase_timestamp'].dt.to_period('M')

# Calculate monthly sales revenue (price * quantity)
monthly_sales = order_data.groupby('order_month')['price'].sum().reset_index()
monthly_sales['order_month'] = monthly_sales['order_month'].dt.to_timestamp()

# Plot
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_sales, x='order_month', y='price', marker='o')
plt.title('Monthly Sales Revenue Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales (BRL)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Translate product categories to English for clarity
category_translation = pd.read_csv('/content/product_category_name_translation.csv')
order_data = order_data.merge(category_translation, on='product_category_name', how='left')

# Aggregate sales by category
category_sales = order_data.groupby('product_category_name_english')['price'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x=category_sales.values, y=category_sales.index, palette='viridis')
plt.title('Top 10 Product Categories by Sales')
plt.xlabel('Total Sales (BRL)')
plt.ylabel('Product Category')
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales by customer state
state_sales = order_data.groupby('customer_state')['price'].sum().sort_values(ascending=False).head(10)

# Plot top 10 states
plt.figure(figsize=(12,6))
sns.barplot(x=state_sales.values, y=state_sales.index, palette='magma')
plt.title('Top 10 Customer States by Sales')
plt.xlabel('Total Sales (BRL)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# Calculate delivery delay in days
order_data['delivery_delay_days'] = (
    order_data['order_delivered_customer_date'] - order_data['order_estimated_delivery_date']
).dt.days

# Remove rows with missing delivery info
delivery_data = order_data.dropna(subset=['delivery_delay_days'])

# Classify as 'On Time' or 'Delayed'
delivery_data['delivery_status'] = delivery_data['delivery_delay_days'].apply(
    lambda x: 'On Time' if x <= 0 else 'Delayed'
)

# Plot the proportion of on-time vs delayed deliveries
plt.figure(figsize=(6,6))
delivery_data['delivery_status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['#2ecc71', '#e74c3c'],
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Delivery Performance: On-Time vs. Delayed Orders')
plt.ylabel('')
plt.show()

# Optional: print quick stats
print(delivery_data['delivery_status'].value_counts(normalize=True) * 100)
```

```python
# Group by seller_id and sum the total sales value
top_sellers = order_data.groupby('seller_id')['price'].sum().sort_values(ascending=False).head(10)

# Plot top 10 sellers
plt.figure(figsize=(12,6))
sns.barplot(
    y=top_sellers.index,
    x=top_sellers.values,
    palette='cool',
    hue=top_sellers.index,
    dodge=False,
    legend=False
)
plt.title('Top 10 Sellers by Revenue')
plt.xlabel('Total Sales (BRL)')
plt.ylabel('Seller ID')
plt.tight_layout()
plt.show()

# Optional: show the data
top_sellers.to_frame('Total_Sales').style.format({'Total_Sales': 'R$ {:,.2f}'})
```

```python
# Select key cleaned columns
final_cols = [
    'order_id',
    'order_purchase_timestamp',
    'price',
    'freight_value',
    'product_category_name_english',
    'customer_state',
    'delivery_delay_days'
]
cleaned_data = order_data[final_cols]

# Export to CSV
cleaned_data.to_csv('cleaned_olist_sales_data.csv', index=False)
print(" Cleaned dataset exported successfully!")
```
